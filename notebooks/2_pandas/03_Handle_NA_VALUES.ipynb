```python
import pandas as pd
```

```python
df = pd.read_csv("data/weather_data.csv",parse_dates=['day'])
```

```python
df.head()
```

```python
df.set_index('day',inplace=True)
```

```python
df.head()
```

```python
df.fillna({'temperature':df.temperature.mean(),
           'windspeed':df.windspeed.mean(),
           'event':'No Event'})
```

```python
df.shape
```

```python
df.fillna(method='ffill',limit=2)
```

```python
df.fillna(method='bfill')
```

```python
# Keep filling the value 
df.interpolate()
```
