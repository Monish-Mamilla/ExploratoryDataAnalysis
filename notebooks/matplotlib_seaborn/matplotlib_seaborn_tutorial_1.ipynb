```python
import pandas as pd

df_sales = pd.read_excel("linechart.xlsx")
df_sales.head()
```

### Line Chart

```python
from matplotlib import pyplot as plt 

plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"], color="blue",label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="orange",label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="green",label="Washing Machine")

plt.title("Product Sales")
plt.xlabel("Revenue (nln $)")
plt.ylabel("Financial Quater")
plt.legend()

plt.show()
```

#### Bar Chart

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index,autopct="%1.2f%%",explode=(0.1,0,0),
        shadow=True,startangle=140)
plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2 = df_sales.set_index("Quarter")
df_sales_2
```

```python
# By Default it takes label as index, df_sales_2 index is Quarter
df_sales_2.plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

### Histogram

```python
import seaborn as sns
```

```python
df_score = pd.read_excel("histograms.xlsx")
df_score.head(3)
```

```python
sns.histplot(data=df_score, x="Exam_Score",kde=True)
plt.xlabel("Score")
plt.ylabel("Count")
plt.title("Exam Scores")
plt.show()
```
